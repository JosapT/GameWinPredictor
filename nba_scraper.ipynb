```python
season_types = ['Regular%20Season', 'Playoffs', 'Pre%20Season', 'All%20Star', 'Finals', 'Playin']
years = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24']


for season_type in season_types:
    for year in years:
        input_url = f'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season={year}&SeasonType={season_type}&StatCategory=PTS'
        req = requests.get(url=input_url).json()

        # Check if the response contains the expected data
        if ('resultSet' not in req) or ('rowSet' not in req['resultSet']):
            print(f"No data found for {year} {season_type}")
            continue
    
        table_headers = req['resultSet']['headers']
        df = pd.DataFrame(req['resultSet']['rowSet'], columns=table_headers)
        df.to_csv(f'/Users/josephthi/Desktop/Personap Projects/Game Winning Predictor/{year}_{season_type}.csv', index=False)
        time.sleep(3)  # Sleep to avoid hitting the API too quickly


```
